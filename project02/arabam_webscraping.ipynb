# Web Scraping

```python
import requests
from bs4 import BeautifulSoup as bts
import pandas as pd
import re
import numpy as np
import time
```

```python
def getAndParseURL(url):
    result = requests.get(url, headers={"User-Agent":"Chrome/115.0.5790.170 "})
    soup = bts(result.text, 'html.parser')
    return soup
```

```python
pages = ["https://www.arabam.com/ikinci-el/otomobil?take=50"]
for page in range(2,51):
    pages.append("https://www.arabam.com/ikinci-el/otomobil?take=50&page=" + str(page))
    
pages
```

```python
links = []

for page in pages:
    html = getAndParseURL(page)
    for rslt in html.findAll("a",{"class":"smallest-text-minus ovh"}):
        links.append('https://www.arabam.com/' + rslt.get('href'))
links
```

```python
len(links)
```

```python
result = []
for rslt in links:
    html = getAndParseURL(rslt)
    
    try:
        price = html.find("div", {"class":"product-properties"}).find(string=re.compile("Fiyat")).findNext().text.replace('TL', '').replace('.', '').strip()
    except:
        price = np.nan
    try:
        make = html.find("div", {"class":"product-properties"}).find(string=re.compile("Marka")).findNext().text.strip()
    except:
        make = np.nan
    try:
        series = html.find("div", {"class":"product-properties"}).find(string=re.compile("Seri")).findNext().text.strip()
    except:
        series = np.nan
    try:
        model = html.find("div", {"class":"product-properties"}).find(string=re.compile("Model")).findNext().text.strip()
    except:
        model = np.nan
    try:
        year = html.find("div", {"class":"product-properties"}).find(string=re.compile("Yl")).findNext().text.strip()
    except:
        year = np.nan
    try:
        km = html.find("div", {"class":"product-properties"}).find(string=re.compile("Kilometre")).findNext().text.replace('km', '').replace('.', '').strip()
    except:
        km = np.nan
    try:
        transmission = html.find("div", {"class":"product-properties"}).find(string=re.compile("Vites Tipi")).findNext().text.strip()
    except:
        transmission = np.nan
    try:
        fuel = html.find("div", {"class":"product-properties"}).find(string=re.compile("Yakt Tipi")).findNext().text.strip()
    except:
        fuel = np.nan
    try:
        body_type = html.find("div", {"class":"product-properties"}).find(string=re.compile("Kasa Tipi")).findNext().text.strip()
    except:
        body_type = np.nan
    try:
        warranty = html.find("div", {"class":"tab-content-car-information"}).find(string=re.compile("Garanti Durumu")).findNext().text.strip()
    except:
        warranty = np.nan
    try:
        from_ = html.find("div", {"class":"product-properties"}).find(string=re.compile("Kimden")).findNext().text.strip()
    except:
        from_ = np.nan
    try:
        wheel_drive = html.find("div", {"class":"product-properties"}).find(string=re.compile("eki")).findNext().text.strip()
    except:
        wheel_drive = np.nan
    try:
        cylinder_number = html.find("div", {"class":"tab-content-car-information"}).find(string=re.compile("Silindir Says")).findNext().text.strip()
    except:
        cylinder_number = np.nan
    try:
        torque = html.find("div", {"class":"tab-content-car-information"}).find(string=re.compile("Tork")).findNext().text.replace('nm', '').strip()
    except:
        torque = np.nan
    try:
        engine_capacity = html.find("div", {"class":"product-properties"}).find(string=re.compile("Motor Hacmi")).findNext().text.replace('cc', '').strip()
    except:
        engine_capacity = np.nan
    try:
        engine_power = html.find("div", {"class":"product-properties"}).find(string=re.compile("Motor Gc")).findNext().text.replace('hp', '').strip()
    except:
        engine_power = np.nan
    try:
        max_power = html.find("div", {"class":"tab-content-car-information"}).find(string=re.compile("Maksimum G")).findNext().text.replace('rpm', '').strip()
    except:
        max_power = np.nan
    try:
        min_power = html.find("div", {"class":"tab-content-car-information"}).find(string=re.compile("Minimum G")).findNext().text.replace('rpm', '').strip()
    except:
        min_power = np.nan
    try:
        acceleration = html.find("div", {"class":"tab-content-car-information"}).find(string=re.compile("Hzlanma")).findNext().text.replace('sn', '').strip()
    except:
        acceleration = np.nan
    try:
        max_speed = html.find("div", {"class":"tab-content-car-information"}).find(string=re.compile("Maksimum Hz")).findNext().text.replace('km/s', '').strip()
    except:
        max_speed = np.nan
    try:
        average_fuel_consumption = html.find("div", {"class":"product-properties"}).find(string=re.compile("Ort. Yakt Tketimi")).findNext().text.replace('lt', '').strip()
    except:
        average_fuel_consumption = np.nan
    try:
        painted_changed = html.find("div", {"class":"product-properties"}).find(string=re.compile("Boya-deien")).findNext().text.strip()
    except:
        painted_changed = np.nan
        
    result.append([price, make, series, model, year, km, transmission, fuel, body_type, warranty, from_, wheel_drive, cylinder_number, torque, engine_capacity, engine_power, max_power, min_power, acceleration, max_speed, average_fuel_consumption, painted_changed])
    
    
columns = ['price_try', 'make', 'series', 'model', 'year', 'km', 'transmission', 'fuel', 'body_type', 'warranty', 'from_', 'wheel_drive', 'cylinder_number', 'torque_nm', 'engine_capacity_cc', 'engine_power_hp', 'max_power_rpm', 'min_power_rpm', 'acceleration_0to100_sec', 'max_speed_kmh', 'average_fuel_consumption_lt', 'painted_changed']
df = pd.DataFrame.from_records(result, columns=columns)
```
